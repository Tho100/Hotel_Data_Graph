```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

plt.style.use('ggplot')

data = pd.read_csv("C:\\Users\\HP\\documents\\Datascience Task\\house_data.csv")

data = data.rename(columns={'bedrooms': 'bedroom number'})

# Bedroom Data
data['bedroom number'].value_counts().plot(kind='bar')
plt.title('Bedroom Data')
plt.xlabel('Bedroom Number')
plt.ylabel('Total Number Of Bedrooms')
plt.show()
# Bedroom Price
plt.bar(data['bedroom number'].head(5),data['price'].head(5))
plt.title('Bedroom Price')
plt.ylabel('Price')
plt.xlabel('Bedroom Number')
plt.show()


plt.scatter(data['grade'].head(5),data['price'].head(5))
plt.title('Price Based On Grade')
plt.ylabel('Price')
plt.xlabel('Grade')

plt.plot([6,11],[0,600000])
plt.show()

data.head()
```
